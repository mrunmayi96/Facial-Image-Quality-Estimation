# Chokepoint Classification 

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
import pandas as pd
from IPython.html import widgets
from IPython.html.widgets import interact
from IPython.display import display
from sklearn.cross_validation import train_test_split
from sklearn import metrics
import numpy as np
from sklearn.metrics import roc_curve, auc
from scipy import interp
import matplotlib.pyplot as plt
from itertools import cycle
```

```python
df = pd.read_csv('./Data/Classification_Dataset.csv')
df
```

```python
X = df[['Brightness', 'FaceArea', 'Sharpness', 'EyesDistance', 'MouthDistance', 'Eyeglass','EyesOpen','Gender', 'Beard', 'Moustache']].values

Y = df[['Quality']].values

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
print X.shape
print Y.shape
```

```python
font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 30}

plt.rc('font', **font)
```

### Naive Bayes

```python
model_nb = GaussianNB()
model_nb.fit(X_train, y_train)
y_pred1 = model_nb.predict(X_test)
print metrics.classification_report(y_test, y_pred1)
print "confusion matrix: \n" , metrics.confusion_matrix(y_test, y_pred1)
print "accuracy score: " , metrics.accuracy_score(y_test, y_pred1)
```

```python
fpr, tpr, threshold = metrics.roc_curve(y_test, y_pred1)
roc_auc = metrics.auc(fpr, tpr)

plt.title('ROC - Gaussian Naive Bayes')
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 16
fig_size[1] = 13
plt.rcParams["figure.figsize"] = fig_size
plt.plot(fpr, tpr, 'cyan', label = 'NB AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right' , fontsize = 25)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])1000
plt.ylabel('True Positive Rate',fontsize=20)
plt.xlabel('False Positive Rate',fontsize=20)
plt.colors
#plt.show()

#plt.savefig('./Images/ROC_NB.eps', format='eps', dpi=1000)
```

### Decision Tree

```python
def DT(num):
    model_dt = DecisionTreeClassifier(max_depth=num)
    model_dt.fit(X_train, y_train)
    y_pred2 = model_dt.predict(X_test)
    print metrics.classification_report(y_test, y_pred2)
    print "confusion matrix: \n" , metrics.confusion_matrix(y_test, y_pred2)
    print "accuracy score: " , metrics.accuracy_score(y_test, y_pred2)
    return y_pred2
```

```python
#i = interact(DT, num=(1,20))
y_pred2 = DT(5)
#Best at 5
```

```python
model_dt = DecisionTreeClassifier(max_depth=5)
model_dt.fit(X_train, y_train)
y_pred2 = model_dt.predict(X_test)
print metrics.classification_report(y_test, y_pred2)
print "confusion matrix: \n" , metrics.confusion_matrix(y_test, y_pred2)
print "accuracy score: " , metrics.accuracy_score(y_test, y_pred2)


fpr, tpr, threshold = metrics.roc_curve(y_test, y_pred2)
roc_auc = metrics.auc(fpr, tpr)

plt.title('ROC - Decision Tree')
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 16
fig_size[1] = 13
plt.rcParams["figure.figsize"] = fig_size
plt.plot(fpr, tpr, 'g', label = 'DT AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right', fontsize = 25)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate',fontsize=20)
plt.xlabel('False Positive Rate',fontsize=20)
#plt.show()

#plt.savefig('./Images/ROC_DT.eps', format='eps', dpi=1000)
```

### Random Forest

```python
def RF(num):
    model_rf = RandomForestClassifier(criterion='entropy',n_estimators=num)
    model_rf.fit(X_train, y_train)
    y_pred3 = model_rf.predict(X_test)
    print metrics.classification_report(y_test, y_pred3)
    print "confusion matrix: \n" , metrics.confusion_matrix(y_test, y_pred3)
    print "accuracy score: " , metrics.accuracy_score(y_test, y_pred3)
```

```python
#i = interact(RF, num=(1,50))
RF(23)
#Best at 23
```

```python
model_rf = RandomForestClassifier(criterion='entropy',n_estimators=23)
model_rf.fit(X_train, y_train)
y_pred3 = model_rf.predict(X_test)
print metrics.classification_report(y_test, y_pred3)
print "confusion matrix: \n" , metrics.confusion_matrix(y_test, y_pred3)
print "accuracy score: " , metrics.accuracy_score(y_test, y_pred3)

fpr, tpr, threshold = metrics.roc_curve(y_test, y_pred3)
roc_auc = metrics.auc(fpr, tpr)

plt.title('ROC - Random Forest')
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 16
fig_size[1] = 13
plt.rcParams["figure.figsize"] = fig_size
plt.plot(fpr, tpr, 'm', label = 'RF AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right' , fontsize = 25)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate',fontsize=20)
plt.xlabel('False Positive Rate',fontsize=20)
#plt.show()

#plt.savefig('./Images/ROC_RF.eps', format='eps', dpi=1000)
```

### SVM

```python
model_svm = svm.SVC(kernel='rbf', gamma = 0.1)
model_svm.fit(X_train, y_train)
y_pred4 = model_svm.predict(X_test)
print metrics.classification_report(y_test, y_pred4)
print "confusion matrix: \n" , metrics.confusion_matrix(y_test, y_pred4)
print "accuracy score: " , metrics.accuracy_score(y_test, y_pred4)

#Gamma = 1/num_features

fpr, tpr, threshold = metrics.roc_curve(y_test, y_pred4)
roc_auc = metrics.auc(fpr, tpr)

plt.title('ROC - SVM')
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 16
fig_size[1] = 13
plt.rcParams["figure.figsize"] = fig_size
plt.plot(fpr, tpr, 'y', label = 'SVM AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right' , fontsize = 25)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate',fontsize=20)
plt.xlabel('False Positive Rate',fontsize=20)
#plt.show()

#plt.savefig('./Images/ROC_SVM.eps', format='eps', dpi=1000)
```
