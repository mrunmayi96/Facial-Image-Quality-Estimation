### Read DL Features CSV

```python
data = pd.read_csv('./Data/CP_DLFeatures.csv',header=None)
print data.shape
data.dropna()
print data.shape
#expected shape = 29022 x 2048
```

```python
scaler = StandardScaler().fit(data)
rescaledX = scaler.transform(data)
np.savetxt("./Data/ScaledX_CP.csv", rescaledX, delimiter=",")
```

```python
X = pd.read_csv('./Data/ScaledX_CP.csv',header=None)
```

### Analyze dimensions

```python
pca = PCA(n_components=2048)
pca.fit(X)
var= pca.explained_variance_ratio_
var1=np.cumsum(np.round(pca.explained_variance_ratio_, decimals=4)*100)

plt.plot(var1)
```

```python
pca = PCA(n_components=1300)
reduced_X = pca.fit_transform(X)
reduced_X.shape
#expected shape = 29022 x 1300
```

```python
np.savetxt('./Data/CP_StdDL.csv', reduced_X, delimiter=',')
```
