```python
import numpy as np
import pandas as pd 
import seaborn as sns
from sklearn.cross_validation import train_test_split
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('./Data/Classification_Dataset.csv')
df
```

```python
X = df[['Brightness', 'FaceArea', 'Sharpness', 'EyesDistance', 'MouthDistance', 'Eyeglass','EyesOpen','Gender', 'Beard', 'Moustache']].values

Y = df[['Quality']].values

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)

X.shape
```

```python
np.random.seed(7)
```

```python
from keras.layers import Dense, Dropout
from keras.models import Sequential
from keras.models import load_model
from pylab import rcParams
from keras.callbacks import ModelCheckpoint, TensorBoard
```

```python
nb_epoch = 1000
batch_size = 100

# create model
model = Sequential()
```

```python
# first layer
model.add(Dense(10, input_dim=10, activation= 'relu' ))
model.add(Dropout(0.2))

# second layer 
model.add(Dense(5, activation= 'relu' ))
model.add(Dropout(0.2))

# third layer 
model.add(Dense(2, activation= 'relu' ))
model.add(Dropout(0.2))

# output layer
model.add(Dense(1, activation= 'sigmoid'))
```

```python
# Compile model
#model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Best Practice
model.compile(optimizer='adam', 
                    loss='binary_crossentropy', 
                    metrics=['accuracy'])

checkpointer = ModelCheckpoint(filepath="./Data/DNN_Best_Model.h5",
                               verbose=0,
                               save_best_only=True)
tensorboard = TensorBoard(log_dir='./logs',
                          histogram_freq=0,
                          write_graph=True,
                          write_images=True)
```

```python
# Fit the model
#model.fit(X_train, y_train, epochs=1000, batch_size=100)

# Best Practice
history = model.fit(X_train, y_train,
                    epochs=nb_epoch,
                    batch_size=batch_size,
                    shuffle=True,
                    validation_data=(X_test, y_test),
                    verbose=1,
                    callbacks=[checkpointer, tensorboard]).history
```

```python
model = load_model('./Data/DNN_Best_Model.h5')

sns.set(style='whitegrid', palette='muted', font_scale=1.5)

rcParams['figure.figsize'] = 14, 8
```

```python
plt.plot(history['loss'])
plt.plot(history['val_loss'])
plt.plot(history['acc'])
plt.title('model evaluation')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right');scores = model.evaluate(X_test, y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
