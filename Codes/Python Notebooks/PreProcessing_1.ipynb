# Facial Image Quality Estimation

### Pre-process captured images

```python
import dlib
from PIL import Image
from skimage import io
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import numpy as np

def detect_faces(image):

    # Create a face detector
    face_detector = dlib.get_frontal_face_detector()

    # Run detector and get bounding boxes of the faces on image.
    detected_faces = face_detector(image, 1)
    face_frames = [(x.left(), x.top(),
                    x.right(), x.bottom()) for x in detected_faces]

    return face_frames

```

```python
import os

for dirName, subDirList, fileList in os.walk('./Data/C1'):
   
    if(len(fileList) > 0):
        fileList.sort()
        
        for img in fileList:
            
            img_path = './Data/C1/'+img
            imgName, ext = img.split('.jpg')
            imgName = imgName + ".png"
            image = io.imread(img_path)
            detected_faces = detect_faces(image)

            for n, face_rect in enumerate(detected_faces):
                face = Image.fromarray(image).crop(face_rect)
                plt.subplot(1, len(detected_faces), n+1)
                plt.axis('off')
                plt.imshow(face)
                plt.savefig('Data/C1_Cropped/'+imgName,transparent = True, bbox_inches = 'tight', pad_inches = 0)

           

```

```python
import os
for dirName, subDirList, fileList in os.walk('./Data/C1_Cropped'):
       if(len(fileList) > 0):
        fileList.sort()
        for img in fileList:
            
            cropped_image = cv2.imread('./Data/C1_Cropped/'+img)
            grayimage =  cv2.cvtColor(cropped_image, cv2.COLOR_BGR2GRAY)
            cv2.imwrite('./Data/C1_Gray/'+img, grayimage)
            
            
```
