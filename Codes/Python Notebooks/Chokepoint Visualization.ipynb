```python
#import statements

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.decomposition import PCA
from io import BytesIO
```

```python
df = pd.read_csv('./Data/foo.csv')

Y = df[['MatchScore']].values

#df
```

## 2. Visualize Dataset

### A) Continuous Value Variables

```python
sns.set(style="whitegrid", color_codes=True)
```

```python
dims = (15, 15)

fig, axs = plt.subplots(nrows=2, ncols=3, figsize=dims)
sns.regplot(x='FaceArea', y='MatchScore', data=df, ax=axs[0][0])
sns.regplot(x='Brightness', y='MatchScore', data=df, ax=axs[0][1])
sns.regplot(x='Sharpness', y='MatchScore', data=df, ax=axs[0][2])

sns.regplot(x='EyesDistance', y='MatchScore', data=df, ax=axs[1][0])
sns.regplot(x='MouthDistance', y='MatchScore', data=df, ax=axs[1][1])
sns.regplot(x='Confidence', y='MatchScore', data=df, ax=axs[1][2])
axs[1,2].set_xlim(0,7)
```

```python
dims = (15, 15)

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=dims)

sns.boxplot(x='EyesOpen', y='MatchScore', data=df, ax=axs[0][0])
sns.boxplot(x='Eyeglass', y='MatchScore', data=df, ax=axs[0][1])
sns.boxplot(x='Moustache', y='MatchScore', data=df, ax=axs[1][0])
sns.boxplot(x='Beard', y='MatchScore', data=df, ax=axs[1][1])
```

```python

sns.set(style='whitegrid', context='notebook')
cols = ['Sharpness', 'Brightness', 'FaceArea', 'Beard', 'Moustache', 'MatchScore']

sns.pairplot(df[cols], size=3)
plt.tight_layout()
plt.show()
```

```python
cols = ['Sharpness', 'Brightness', 'FaceArea', 'Beard', 'Moustache', 'MatchScore']

cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.5)
hm = sns.heatmap(cm,
                 cbar=True,
                 annot=True,
                 square=True,
                 fmt='.2f',
                 annot_kws={'size': 15},
                 yticklabels=cols,
                 xticklabels=cols)

plt.show()
```
